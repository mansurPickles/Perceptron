```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def char2vec(char):
    if char == '.':
        return -1
    else:
        return 1

    
def convertChar(X: list):
    X1 = [[0]*len((X[0]))]*len(X)
    xtemp = list()
    X2 = list(list())
    
    for i in range(len(X)):
        for j in range(len(X[0])):
            c = X[i][j]


            if X[i][j] == ".":
                xtemp.append(0)
            elif X[i][j] == "#":
                xtemp.append(1)


        temp = xtemp.copy()
        xtemp.clear()
        X2.append(temp)
    

    return X2
```

```python
def threshold(y: int):
    if y >= 0:
        return 1
    else:
        return 0
```

```python
def init_weights(X:list) -> list:
    weights = list()
    index = 0
    for i in range(len(X)):
        for j in range(len(X[0])):
            weights.insert(index, random.uniform(0,1))
            index += 1
    return weights
```

```python
def predict(X:list, weights:list, b: int):
    y_pred = (np.dot(weights.T, X) + b)
    y_pred = threshold(y_pred)
    return y_pred
```

```python
#init weights

X = np.array(TRAINING_DATA)
weights = []
for i in X:
    a = init_weights(i)
    weights.append(a)
    
weights = np.array(weights)
```

```python
#init x and convert to 0/1
x1 = []

for i in X:
    a = convertChar(i)
    x1.append(a.copy())
    
x = np.array(x1)
x.shape
x = x.reshape(26,35)
```

```python
learning_rate = .03
```

```python
#making bias list

b = []

for i in range(26):
    b.append(random.random())
```

```python
#init y
y = np.identity(26)
```

```python
#this one works, but b doesnt get updated. I realized that this is passing in an int and not the entire 
#list which is why its not being updated, but when I went to fix that it still wouldnt work. 
#This is the closest i've gotten

def learning_algh(neuron_num: int, b: list, epoch_count):
    
    #how many epochs
    for epoch in range(epoch_count):
        
        for i in range(len(x)):
            #prediction    
            pred = predict(x[i], weights[neuron_num], b[neuron_num])
            #error
            error = (y[i][neuron_num] - pred)
            
            #update weights
            for j in range(35):
                weights[neuron_num][j] += (learning_rate * error * x[i][j])
                
            #update bias
            b[neuron_num] += (learning_rate * error)

            
```

```python
#train model

for i in range(26):
    learning_algh(i, b, 100)
    print(f'training neuron: {i} ')
```

```python
#evaluating training data

for i in range(26):
    A2 = 'A'
    print(f'neuron: {chr(ord(A2)+i)}')
    for j in range(26):
        A1 = 'A'
        pred = predict(x[i],weights[j],b[j])
        str1 = f'{chr(ord(A1)+j)} : {pred}'
        print(str1)
    print()
```
