```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def char2vec(char):
    if char == '.':
        return -1
    else:
        return 1

    
def convertChar(X: list):
    X1 = [[0]*len((X[0]))]*len(X)
    xtemp = list()
    X2 = list(list())
    
    for i in range(len(X)):
        for j in range(len(X[0])):
            c = X[i][j]


            if X[i][j] == ".":
                xtemp.append(0)
            elif X[i][j] == "#":
                xtemp.append(1)


        temp = xtemp.copy()
        xtemp.clear()
        X2.append(temp)
    

    return X2
```

```python
def threshold(y: int):
    if y >= 0:
        return 1
    else:
        return 0
```

```python
def init_weights(X:list) -> list:
    weights = list()
    index = 0
    for i in range(len(X)):
        for j in range(len(X[0])):
            weights.insert(index, random.uniform(0,1))
            index += 1
    return weights
```

```python
def predict(X:list, weights:list, b: int):
    y_pred = (np.dot(weights.T, X) + b)
    y_pred = threshold(y_pred)
    return y_pred
```
